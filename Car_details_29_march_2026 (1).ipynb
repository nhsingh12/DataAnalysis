```python
def remove_char(x):
    kmpl, cc, bhp = x
    kmpl = kmpl.split(' ')[0]
    cc = cc.split(' ')[0]
    bhp = bhp.split(' ')[0]

    return kmpl, cc, bhp
```

```python
cars[['mileage(kmpl)', 'engine(CC)', 'max_power(bhp)']] = cars[['mileage', 'engine', 'max_power']].apply(remove_char, axis = 1, result_type = 'expand')
```

```python
cars.head()
```

```python
cars.drop(columns = ['mileage', 'engine', 'max_power'], inplace = True)
```

```python
cars.head()
```

```python
cat_cols = ['fuel']
dummies = pd.get_dummies(cars[cat_cols], drop_first = True)
```

```python
dummies
```

```python
cars = pd.concat([cars, dummies], axis = 1)
```
