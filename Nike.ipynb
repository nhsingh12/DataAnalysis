```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6, 10))
ax1.bar(df1['Year'], df1['Open'], color='blue')
ax1.set_title('Yearly Open Price')
ax1.tick_params(axis='x', rotation=90)  # Rotates labels 90 degrees

# --- Plot 2 ---
ax2.bar(df1['Year'], df1['Close'], color='green')
ax2.set_title('Yearly Close Price')
ax2.tick_params(axis='x', rotation=90)

# --- Plot 3 ---
ax3.bar(df1['Year'], df1['High'], color='red')
ax3.set_title('Yearly High Price')
ax3.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
